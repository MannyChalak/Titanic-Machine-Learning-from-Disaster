```python
import numpy as np
import pandas as pd
```

```python
train_ds = pd.read_csv("train.csv")
train_ds
```

```python
train_ds = train_ds[["Survived", "Pclass", "Name", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]]
train_ds
```

```python
train_ds = train_ds.dropna()
train_ds
```

```python
train_ds.describe()
```

```python
from tensorflow.keras.utils import to_categorical
Pclass = train_ds["Pclass"].to_numpy()
as_categorical_pclass = to_categorical(Pclass - 1, num_classes = 3)
as_categorical_pclass
```

```python
def sex_str_to_bin(s):
    if s=="male":
        return 1
    else:
        return 0

sex = train_ds["Sex"]
as_bin_sex = sex.apply(sex_str_to_bin)
as_bin_sex
```

```python
age_np = train_ds["Age"].to_numpy()
age_np = (train_ds["Age"] - train_ds["Age"].mean()) / train_ds["Age"].std()
age_np[:10]
```

```python
sibsp = train_ds["SibSp"]
sibsp_np = sibsp.to_numpy()
sibsp_np = (train_ds["SibSp"] - train_ds["SibSp"].mean()) / train_ds["SibSp"].std()
sibsp_np[:10]
```

```python
parch = train_ds["Parch"]
parch_np = parch.to_numpy()
parch_np = (train_ds["Parch"] - train_ds["Parch"].mean()) / train_ds["Parch"].std()
parch_np[:10]
```

```python
fare = train_ds["Fare"]
fare_np = fare.to_numpy()
fare_np = (train_ds["Fare"] - train_ds["Fare"].mean()) / train_ds["Fare"].std()
fare_np[:10]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
X_train = np.concatenate((as_categorical_pclass,
                as_bin_sex.values.reshape(-1,1),
                age_np.values.reshape(-1,1),
                sibsp_np.values.reshape(-1,1),
                parch_np.values.reshape(-1,1),
                fare_np.values.reshape(-1,1)), axis=1)
X_train
```

```python
Y_train = train_ds["Survived"].to_numpy()
Y_train[:10]
```

```python
model = LogisticRegression(random_state = 0).fit(X_train, Y_train)
model.score(X_train, Y_train)
```

```python
test_ds = pd.read_csv("test.csv")
test_ds
```

```python
test_ds = test_ds[['PassengerId', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
test_ds
```

```python
test_ds = test_ds.fillna(test_ds['Age'].mean())
len(test_ds)
```

```python
pclass = test_ds['Pclass'].to_numpy()
as_categorical_pclass = to_categorical(pclass - 1, num_classes=3)
as_categorical_pclass[:10]
```

```python
sex = test_ds['Sex']
as_binary_sex = sex.apply(sex_str_to_bin).to_numpy()
as_binary_sex[:10]
```

```python
age_np = test_ds['Age'].to_numpy()
age_np = (age_np - train_ds['Age'].mean()) / train_ds['Age'].std()
age_np[:10]
```

```python
sibsp = test_ds['SibSp']
sibsp_np = sibsp.to_numpy()
sibsp_np = (sibsp_np - train_ds['SibSp'].mean()) / train_ds['SibSp'].std()
sibsp_np[:10]
```

```python
parch = test_ds['Parch']
parch = parch.to_numpy()
parch = (parch - train_ds['Parch'].mean()) / train_ds['Parch'].std()
parch[:10]
```

```python
fare = test_ds['Fare'].to_numpy()
fare = (fare - train_ds['Fare'].mean()) / train_ds['Fare'].std()
fare[:10]
```

```python
test_X = np.concatenate((as_categorical_pclass, 
                as_binary_sex.reshape(-1, 1),
                age_np.reshape(-1, 1),
                sibsp_np.reshape(-1,1),
                parch.reshape(-1,1),
                fare.reshape(-1,1)), axis=1)
test_X
```

```python
predictions = model.predict(test_X)
passenger_ids = test_ds['PassengerId'].to_numpy()
sub_df = pd.DataFrame({'PassengerId':passenger_ids, 'Survived':predictions})
```

```python
sub_df.to_csv('submission.csv')
```
